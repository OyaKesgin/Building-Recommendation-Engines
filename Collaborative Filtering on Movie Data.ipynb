```python
user_rating_table=ur.pivot_table(index='userId', columns='title', values='rating')
user_rating_pivot.head()

```

```python
# Filling the missing Values. Finding a way to deal with NAN values. We cant replace with zero
# one alterbative- centre each user rating around zero by dedcuting row average
# then filling missing values with zero.
# this means the missing data is replaced with neutral scores
```

```python
avg_ratings=user_rating_pivot.mean(axis=1)
user_rating_pivot= user_rating_pivot.sub(avg_ratings, axis=0)
print(user_rating_pivot)
```

```python
b=user_rating_pivot.fillna(0)
```

```python
# Finding Similarities
# which matrix prefered will be discussed later.
movie_ratings_centered=b.T
movie_ratings_centered
```
